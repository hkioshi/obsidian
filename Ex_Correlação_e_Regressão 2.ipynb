# Caso 1

1. Faça a correlação de Pearson. Qual interpretação do resultado?
2. Plote o gráfico. Interprete os resultados
3. Faça a regressão. Tem sentido? Calcule y para x = 110
4. Calcule o erro quadrático médio

```python
from numpy.random import randn
from numpy.random import seed
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from scipy import stats


x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

corr, pvalor = pearsonr(x, y)
print('Correlação de Person: %.3f' % corr)
print('P-valor: %.3f' % pvalor)

descida, intercepcao, r, p, std_err = stats.linregress(x, y)

def reta(x):
  return descida * x + intercepcao

modelo = list(map(reta, x))

print(reta(110))

plt.scatter(x, y)
plt.plot(x, modelo)

plt.show()
```

Correlação de Pearson de 0,013: Indica quase nenhuma relação linear entre as variáveis.

P-valor de 0,956: Mostra que a correlação é estatisticamente insignificante, provavelmente ocorrendo por acaso.

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(modelo, y))
```

O erro é muito grande
